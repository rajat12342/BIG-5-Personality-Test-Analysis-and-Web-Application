```python
#Create combined scores columns for each personality category
import pandas as pd
import numpy as np

df = pd.read_csv("cleaned_data.csv")
index = df.columns.get_loc('O10')

category_questions = {'Extraversion': (7,17), 'Neuroticism': (17,27), 'Agreeableness': (27,37), 'Conscientiousness': (37,47), 'Openness': (47,57) }




def sum_question_scores(category_questions):

    for category, (begin_idx, end_idx) in category_questions.items():
        df[category] = df.iloc[:, begin_idx:end_idx].sum(axis=1)

    return df

df = sum_question_scores(category_questions)


df

```

##### The total scores within each of the 5 categories appear to be normally distributed since the median and mean are very similar

```python
df.describe()

#The total scores within each of the 5 categories appear to be normally distributed since the median and mean are very similar
```

The histograms show that responses for all of the 5 categories are mostly normally distributed. However, some categories like openness have the distribution
shifted to the right indicating our respondents might be more open to new experiences than the general population. We can verify this for all of the
5 categories by conducting a one sample t-test with the theoretical mean of 30 for each category, as that's what the literature shows to be the population mean
for all of the 5 aspects of personality.

```python
import matplotlib.pyplot as plt
import seaborn as sns
trait_group = df.iloc[:, 58:63]

for trait in trait_group:
    mean_value = df[trait].mean()
    std_value = df[trait].std()
    
    plt.axvline(mean_value, color='r', linestyle='--', label=f'Mean: {mean_value:.2f}')
    plt.axvline(mean_value + std_value, color='g', linestyle='--', label=f'+1 SD: {mean_value + std_value:.2f}')
    plt.axvline(mean_value - std_value, color='g', linestyle='--', label=f'-1 SD: {mean_value - std_value:.2f}')

    plt.legend()
    plt.title(f'Histogram for {trait}')
    
    sns.histplot(df[trait] , kde= True, bins =30)
    plt.show()
```

```python
from scipy.stats import ttest_1samp

population_mean = 30

for trait in trait_group:
    t_stat, p_value = ttest_1samp(df[trait], population_mean)
    print(f"T-statistic: {t_stat}, p-value: {p_value} for trait: {trait}")
```

It appears that our respondents significantly differ in all aspects of personality compared to the general population. However, since our sample size
is so large these significant test results may not have much practical value since even small differences in the means can produce statistically significant
results due to the large sample size. Hence, we need to check how meaningful and practically significant the differences are by quantifying them

```python
def cohens_d(sample_mean, population_mean, sample_std):
    return (sample_mean - population_mean) / sample_std


for trait in trait_group:
    sample_mean = df[trait].mean()
    sample_std = df[trait].std()
    d = cohens_d(sample_mean, population_mean, sample_std)
    print(f"Cohen's d for {trait}: {d}")
```

0.2: Small effect
0.5: Medium effect
0.8: Large effect

Using the above criteria to interpret cohen's d for each trait, we can conclude that the following:
1) Our sample of respondents are slightly more extroverted than the general population. However the difference is close to being trivial.
2) Our sample of respondents are not more or less neurotic than the general population in any practical sense.
3) Our sample of respondents are moderately more agreeable than the general population.
4) Our sample of respondents are somewhat more conscientious than the general population.
5) Our sample of respondents are greatly more open to new experiences than the general population.

The above results are expected for a sample of respondents who took the personality test on the internet. People who search and take psychological tests
online tend to be more open to new experiences compared to the general population.

Shapiro-Wilk test shows most of the distributions of each race on each of the 5 categories aren't normal. However this isn't very helpful since the
sample size is so large that even small deviations produce a small p value thus indicating a deviation from normality.

```python
import scipy.stats as stats

traits = df.iloc[:,58:63]
traits = traits.columns

for trait in traits:
    print(f"Normality test(shapiro wilk) for trait: {trait}")
    
    for race in df['race_categories'].unique():
        race_data = (df[df['race_categories'] == race][trait])
        test_statistic, p_value = stats.shapiro(race_data)
        print(f"Race : {race}, p_value : {p_value:.2f}")
        if p_value < 0.05:
            print(f"  ->  Data for race {race} is not normally distributed (reject H0)")
        else:
            print(f"  ->  Data for race {race} is normally distributed (do not reject H0)")


        
```

The Q-Q plots show the distributions to be mostly normal with some significant outliers in a decent number of them. Therefore, using a non-parametric test 
like Kruskal-Wallis is preferred in order to check if there are significant differences between the races on the 5 categories.

```python
for trait in traits:
    for race in df['race_categories'].unique():
        race_data = df[df['race_categories'] == race][trait] 
        
        if len(race_data) > 1:
            plt.figure(figsize=(4, 3)) 
            stats.probplot(race_data, dist="norm", plot=plt)
            plt.title(f"Q-Q Plot for {trait} (Race: {race})")
            plt.savefig(f'distribution for {race}{trait}', dpi=300, bbox_inches='tight')
            plt.show()
            
            
        else:
            print(f"Not enough data for {race} in {trait} to create a Q-Q plot")   
```
